```python
params_data = cleaned_data.drop(['Unnamed: 0', 'year'], axis=1)
```

```python
params_data # just the parameters without time info
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(params_data))
```

```python
scaled_data.isna().sum().sum() # there are 2 na items
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
scaled_data = pd.DataFrame(imp.fit_transform(scaled_data))
```

```python
scaled_data.isna().sum().sum() # there are no more na items
```

```python
import seaborn as sns  # draws pretty things 🥳
corr = scaled_data.corr()
# TO-DO : find any beautify extensions for jupyter

# https://towardsdatascience.com/better-heatmaps-and-correlation-matrix-plots-in-python-41445d0f2bec
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 200, n=20, center = 'light'),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
# just to check correlation between different elements
```

```python
# Now we have to create train and test sets
train = pd.concat([scaled_data[start:start+9] for start in range(0,scaled_data.shape[0], 10)])
test = pd.DataFrame([scaled_data.iloc[start+9] for start in range(0,scaled_data.shape[0], 10)])
```

```python
x_train, y_train = train.iloc[:, 0:8] ,train.iloc[:, 8:]
x_test, y_test = test.iloc[:, 0:8] ,test.iloc[:, 8:]
```

```python
train.shape, test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_pred, y_test) ** 0.5
```

```python
rmse
```

```python
from sklearn.kernel_ridge import KernelRidge
kern_ridge = KernelRidge().fit(x_train, y_train)
```

```python
y_pred_ridge = kern_ridge.predict(x_test)
rmse_ridge = mean_squared_error(y_pred_ridge, y_pred)**0.5
```
