```python
import numpy as np
# STANDARDIZATION 
# so the 64 pixels are the features
# standardize each pixel
# for row wise selection transpose
def standardise(data):
    data_transpose = digits.data.T
    for i, row in enumerate(data_transpose):
        if np.std(row) != 0: # using Z score method 
            data_transpose[i] = (row - np.mean(row))/np.std(row)
    return data_transpose.T
```

```python
digits.data = standardise(digits.data)
```

```python
digits.data.shape[1]
```

```python
def cor_matrix(data):
    return (np.matmul(data.T,data))/(data.shape[0] - 1) 
    # don't need to substract the mean, the mean of the data is zero
    # here N (data.shape[0]) stands for the number of instances not the number of features
```

```python
covar_mat = cor_matrix(digits.data)
```

```python
transposed = digits.data.T
help(digits.data.T)
```

```python
import numpy as np
Covariance_matrix=np.cov(transposed)
```

```python
print(Covariance_matrix.shape)
```

```python
np.square(np.subtract(covar_mat, Covariance_matrix)).mean() # they're the same
```

```python
import numpy as np
value,vector=np.linalg.eig(Covariance_matrix)
```

```python
all(value[i] >= value[i+1] for i in range(len(value) - 1))
```

```python
vector = vector[np.argsort(value)]
```

```python
# so our data so far has 64 dimensions/features
# now we need to perform some dimensionality reduction
```

```python
# what to keep and what not to keep
percentage_values = value / sum(value)
# CAN WE TELL FROM THIS LIKE WHICH PIXEL WAS PRESERVED THE MOST OR SOMETHING ?
# don't need to sort them they are already in decreasing order
percentage_values
```

```python
# we want to keep enough dimensions to perserve around atleast 90% of the data 
```

```python
vectors_kept = 0
data_restored = 0

for index, percent in enumerate(percentage_values):
    if data_restored > 0.9: # enough dimensions 
        break
    data_restored += percent
    vectors_kept += 1
```

```python
vectors_kept 
```

```python
# so we've reduced dimensionality from 64 to 31 
# now just transfrom the data
# how to proceed from here ?
```

```python
# CAN WE TELL FROM THIS LIKE WHICH PIXEL WAS PRESERVED THE MOST OR SOMETHING ?
# HOW TO SUMMARIZE NUMPY ARRAYS LIKE DATAFRAMES
# HOW TO CONVERT ONE TO ANOTHER
# HOW MUCH DATA IS ENOUGH DATA ?
```

```python
data_kept = vector[:vectors_kept]
```
